```python
%matplotlib inline

from matplotlib import pyplot
from matplotlib.image import imread
from glob import glob
import os

def plot_image_vs_predictions(imagepath):
    image = imread(imagepath)
    filename, ext = os.path.splitext(imagepath)
    prediction_images = glob(os.path.join(os.path.dirname(imagepath), 'marrnet-predictions', 
                                          os.path.basename(filename) + '*' + ext))
    prediction_images = [imread(prediction_image) for prediction_image in prediction_images]
    fig, axes = pyplot.subplots(1, 1 + len(prediction_images))
    axes[0].imshow(image)
    for ax, prediction_image in zip(axes[1:], prediction_images):
        ax.imshow(prediction_image)
    for ax in axes:
        ax.axis('off')
```

```python
image_directory = '/mnt/c/Users/Martin/workspaces/dicarlo/decomposition/images/sorted/Chairs/_001/V6'
for imagepath in glob(os.path.join(image_directory, '*.png')):
    plot_image_vs_predictions(imagepath)
```
